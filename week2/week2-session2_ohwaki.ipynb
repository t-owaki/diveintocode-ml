### Week2授業前課題2  ダミーデータの作成

```python
'''
モジュールのインポート
'''
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 【問題1】乱数の作成

```python
'''
２次元の正規分布による乱数５００個作成するコード
'''
np.random.seed(0)

mu1 = [-3, 0] # 平均
cov = [[1.0, 0.8],[0.8, 1.0]] # 共分散

#500はデータ数
values = np.random.multivariate_normal(mu1, cov, 500)
print(values.shape)
print(values[0]) # 値が変化していないことを確認
```

#### 【問題2】散布図による可視化

```python
'''
問題1で作成したデータ点を散布図により可視化するコード
'''
x1, x2 = np.random.multivariate_normal(mu1, cov, 500).T

plt.scatter(x1, x2)
plt.title("scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

#### 【問題3】ヒストグラムによる可視化

```python
'''
問題1で作成したデータをヒストグラムにより可視化するコード
'''
plt.hist(x1)
plt.xlim(-7, 4)
plt.show()

plt.hist(x2)
plt.xlim(-7, 4)
plt.show()
```

#### 【問題4】データの追加

```python
'''
問題1に加えてたに平均が(0, -3)、共分散行列が以下で表される2次元正規分布
による乱数を500個作成するコード
'''
np.random.seed(0)

mu1 = [-3, 0]
mu2 = [0, -3]
cov = [[1.0, 0.8],[0.8, 1.0]] # 共分散

#500はデータ数
x1, y1 = np.random.multivariate_normal(mu1, cov, 500).T
x2, y2 = np.random.multivariate_normal(mu2, cov, 500).T

plt.scatter(x1, y1, label="0")
plt.scatter(x2, y2, label="1")
plt.title("scatter plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

#### 【問題5】データの結合

```python
'''
題1、4で作成したndarrayを (1000, 2)に結合するコード
'''
x = np.random.multivariate_normal(mu1, cov, 500)
y = np.random.multivariate_normal(mu2, cov, 500)
print(x.shape, y.shape)

concat_xy = np.concatenate([x, y], axis=0)

print(concat_xy.shape)
print(concat_xy[0])
print(concat_xy[999])
```

#### 【問題6】ラベル付け

```python
'''
問題1、4どちらで作成したものなのかを示す ラベル 付けするコード
(ラベル　問題1：0　問題4:1)
'''
# ラベルの用意
label = [0]*500 + [1]*500
print("label[0] : {} label[500] : {} label個数 : {}".format(label[0], label[500], len(label)))

# データの前処理
data = np.vstack((concat_xy.T, label)).T
print(data.shape)
print(data[0])
print(data[499])
print(data[500])
print(data[999])
```
